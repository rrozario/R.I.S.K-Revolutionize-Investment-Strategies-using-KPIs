```python
dff = pd.read_csv("output.csv")
final = pd.read_csv("tableau.csv")
```

```python
def predict(fb):
    fb = fb.rename(columns={'Date': 'ds', 'High': 'y'})
    # Make the prophet model and fit on the data
    gm_prophet = fbprophet.Prophet(changepoint_prior_scale=0.15,yearly_seasonality=True,daily_seasonality=True )
    gm_prophet.fit(fb)
    gm_forecast = gm_prophet.make_future_dataframe(periods=365 * 2, freq='D')
    # Make predictions
    gm_forecast = gm_prophet.predict(gm_forecast)
#     gm_prophet.plot(gm_forecast)
    return gm_forecast
```

```python
import fbprophet

prophet = pd.DataFrame(columns=["ds","yhat", 'Name'])

for i in range(0,len(final.Name.unique())):#define the ticker symbol
    namstr = final.Name.unique()[i]
    fb = final[final["Name"]==namstr]
    fb = fb[["Date","High"]]
    count = fb["High"].isna().sum()
    ret = predict(fb)
    ret = ret[["ds","yhat"]]
    ret["Name"] = dff["Name"][i]
    ret.reset_index(level=0,inplace=True)
    prophet = prophet.append(ret, ignore_index = True) 
    prophet.to_csv("prophet.csv", sep=',', index=False)
```
